### Feature Scaling

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
```

```python
df =pd.read_csv("student_final.csv")
```

```python
df.head()
```

```python
numerical = [
    "age",
    "Medu",
    "Fedu",
    "traveltime",
    "studytime",
    "failures",
    "famrel",
    "freetime",
    "goout",
    "Dalc",
    "Walc",
    "health",
    "absences",
    "G1",
    "G2",
    "G3"
]
```

```python
df[numerical].head()
```

#### Apply StandardScaler

```python
scaler = StandardScaler()
df_standard = df.copy()
df_standard[numerical] = scaler.fit_transform(df_standard[numerical])
```

```python
df_standard.head()
```

#### Apply MinMaxScaler

```python
minmax = MinMaxScaler()
df_minmax = df.copy()
df_minmax[numerical] = minmax.fit_transform(df_minmax[numerical])
```

```python
df_minmax.head()
```

#### Compare Statistics

##### Original Data

```python
df[numerical].describe()
```

##### Standardized Data

```python
df_standard[numerical].describe()
```

##### Normalized Data

```python
df_minmax[numerical].describe()
```

```python
df_standard.to_csv("student_standardized.csv", index=False)

df_minmax.to_csv("student_normalized.csv", index=False)
```

```python
import matplotlib.pyplot as plt

plt.hist(df["absences"])
plt.title("Original Absences")
plt.show()
```

#### Observation

The original absences data is unevenly distributed. Most students have few or no absences, while only a small number of students have many absences. This results in a right-skewed distribution with a few extreme values (outliers).

```python
plt.hist(df_standard["absences"])
plt.title("Standardized Absences")
plt.show()
```

#### Observation

After applying StandardScaler, the shape of the distribution remains the same, but the values are centered around 0 with a standard deviation close to 1. Standardization changes the scale of the data but does not change its distribution.

```python
plt.hist(df_minmax["absences"])
plt.title("Normalized Absences")
plt.show()
```

#### Observation

After applying MinMaxScaler, all values are transformed to the range 0 to 1. The overall distribution remains the same, but the scale becomes consistent, making it easier for algorithms that require features to be within a fixed range.
