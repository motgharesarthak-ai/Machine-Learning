### Data Visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df =pd.read_csv("student_data.csv")
```

```python
df.head()
```

```python
plt.hist(df["G3"])
plt.title("Distribution of Final Grades")
plt.xlabel("Final Grades")
plt.ylabel("Number of Students")
plt.show()
```

#### Observation

The distribution of final grades (G3) shows that most students scored in the middle range, while very few students achieved extremely high or extremely low marks. This indicates that the majority of students performed at an average level.

```python
sns.countplot(x ="sex", data=df)
plt.title("Male vs Female Students")
plt.show()
```

#### Observation

The count plot shows the number of male and female students in the dataset. The dataset appears to have a fairly balanced distribution of both genders, making it suitable for comparing academic performance between male and female students.

```python
sns.countplot(x="studytime", data=df)
plt.title("Study Time Distribution")
plt.show()
```

#### Observation

Most students belong to the lower study-time categories, while fewer students spend a large amount of time studying each week. This suggests that moderate study time is more common among the students.

```python
sns.boxplot(y="G3", data=df)
plt.title("Final Grades")
plt.show()
```

#### Observation

The box plot summarizes the distribution of final grades and helps identify outliers. Most students' grades fall within a specific range, while a few students scored significantly higher or lower than the rest. These extreme values are potential outliers.

```python
sns.scatterplot(x="absences",y="G3", data =df)
plt.title("Absences vs Final Grades")
plt.show()
```

#### Observation

The scatter plot shows the relationship between absences and final grades. Students with fewer absences generally tend to achieve better grades, although the relationship is not perfectly linear. Some students perform well despite having many absences, indicating that other factors also influence academic performance.

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

#### Observation

The heatmap displays the correlation between numerical features. Variables with values close to 1 have a strong positive correlation, while values close to -1 indicate a strong negative correlation. The previous exam scores (G1 and G2) are expected to have the strongest positive correlation with the final grade (G3), making them important features for prediction.

#### Conclusion

- Most students achieved average final grades.
- Previous exam scores (G1 and G2) appear to be strong predictors of the final grade.
- Students with fewer absences generally perform better.
- Study time seems to have a positive relationship with academic performance.
- The dataset is suitable for building a machine learning model to predict students' final grades.
