### Train Test Split

```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("student_final.csv")
```

```python
df.head()
```

```python
x = df.drop("G3", axis =1)
```

```python
y = df["G3"]
```

```python
print(x.shape)
```

```python
print(y.shape)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2,random_state=42)
```

```python
print("x_train:", x_train.shape)
```

```python
print("x_test:", x_test.shape)
```

```python
print("y_train:", y_train.shape)
```

```python
print("y_test:", y_test.shape)
```

```python
print(x_train.head())
```

```python
print(y_train.head())
```

```python
print("Training Data:", len(x_train))
print("Testing Data:", len(x_test))
```

```python
training_percentage = (len(x_train)/ len(df))* 100
print("Training Percentage:", training_percentage, "%")
```

```python
testing_percentage = (len(x_test)/ len(df))* 100
print("Testing Percentage:", testing_percentage, "%")
```
